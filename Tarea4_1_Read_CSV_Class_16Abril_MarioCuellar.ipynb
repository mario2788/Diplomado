```python
import pandas as pd

class DataReader:
    def __init__(self, file_path):
        self.file_path = file_path
        self.df = None
        self.next_index = 0
        self.data = True
        
    def available_data(self):
        return self.data
    
    def read_data(self):
        try:
            self.df = pd.read_csv(self.file_path)
            return True
        except FileNotFoundError:
            self.data = False
            print(f"Error: File '{self.file_path}' no encontrado.")
            return False
        
    def next_data(self):
        if self.df is not None and self.next_index < len(self.df)-1:
            data = self.df.iloc[self.next_index:self.next_index+2]
            self.next_index += 2
            return data
        else:
            self.data = False
            print("No data")
            return False

# Crear instancia de la clase DataReader
reader = DataReader("iris.csv")

# Leer el archivo de datos
if reader.read_data():
    
    # Imprimir el DataFrame interno
    print(reader.df)
    
    # Imprimir pares de datos
    while(True):
        a = reader.next_data()
        if( not reader.available_data() ):
            break
        print(a)
    
```
